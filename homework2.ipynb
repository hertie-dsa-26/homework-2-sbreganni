## Question 3: Algorithm Analysis (2 points)

Consider the following algorithm, which finds the contiguous subarray with the largest sum:

```python
max_sum = None
for i in range(len(array)):
    sum_subarray = 0
    for j in range(i, len(array)):
        sum_subarray += array[j]
        if max_sum is None or max_sum < sum_subarray:
            max_sum = sum_subarray
print(max_sum)
```

Analyze the efficiency of this algorithm:

1. Is this algorithm polynomial? Justify your answer.
2. What is the runtime of this algorithm in **Big-O** notation? Show your reasoning.

*Yes, it is polynomial: the algorithm has nested loops that scale based on the size of the input*

*The runtime of this algorithm is `O(n^2)` (Quadratic Time)*

- **Outer Loop:** *Runs n times (from i = 0 to n-1). This loop is `O(n)`*
- **Inner Loop:** *Runs from the current index i to the end of the array for every iteration of the outer loop. (When i=0, the inner loop runs n times and when i=1, it runs n-1 times and so on.) Ignoring the coefficients in Big O notation this loop is `O(n)`*
- `O(n)` x `O(n)` = `O(n^2)`
